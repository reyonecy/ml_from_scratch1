## Solutions Applied
 - Replaced missing values NaN and N/a with 'Unknown'
 - Removed duplicate data for Costumer_ID 1020
 - Removed unwanted leading and trailing signs and symbols in Last_Name  using str.strip()
 - Properly Formatted the phone numbers in Phone_Number column using a function().
 - Replaced Yes and No in the columns Paying Customer and Do_Not_Contact with 'Y' and 'N' respectively.
 -Removed the unwanted column called Not_Useful_Column.

```python
df.fillna("Unknown",inplace=True)
```

```python
df
```

```python
df.replace("N/a","Unknown", inplace=True)
df
```

```python
df.duplicated()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('.!@#$%^&*()_-+=~`<>?/\\|[]{}"')
```

```python
df
```

```python
df
```

```python
import re

def format_phone_number(phone):
    # Check if the phone number is not an integer. If it is integer (Unknown) return it as is.
    if type(phone) == int:
      return phone

    # Otherwise, format it or return "Invalid Number"
    digits = re.sub(r'\D','',str(phone)) # Convert to string before using re.sub

    if len(digits) == 10:
        return f"{digits[:3]}-{digits[3:6]}-{digits[6:]}"
    else:
        return "Unknown" # Corrected: Return the string "Invalid Number"

df['Phone_Number'] = df['Phone_Number'].apply(format_phone_number)
```

```python
df
```

```python
df['Paying Customer'] = df['Paying Customer'].replace({'Yes': 'Y', 'No': 'N'})
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace({'Yes': 'Y', 'No': 'N'})
```

```python
df
```

```python
df.drop('Not_Useful_Column', inplace=True, axis=1)
```
