```python
# identify null values
isnull = df.isnull()
isnull
```

```python
# find total number of null values
isnull_count = isnull.sum()
isnull_count
```

```python
# total nulls in dataset
total_isnull_count = isnull.sum().sum()
total_isnull_count
```

```python
# Replace single quote wuth blank and make the data formated in date form.
df['Date'] = df["Date"].replace("'","",regex=True)
df
```

```python
# handle null values , return dataframe with no empty cells.
# we can use df.dropna(), it removes empty cells but will not change the dataframe.
# if we want data frame to be changed, simply use inplace = True
new_df = df.dropna()
new_df
```

```python
isnull_count_newdf = new_df.isnull()
isnull_count_newdf
```

```python
# another way of handling the null values by replacing the null values
df.fillna(130,inplace=True)
df
```

```python
df = pd.read_csv("./sample_data/pandas_dataset.csv")
df['Calories'].fillna(130,inplace=True)
```

```python
# fill specific with mean , median and mode
df = pd.read_csv("./sample_data/pandas_dataset.csv")
x = df["Calories"].mean()
df["Calories"].fillna(x, inplace = True)
df
```
